```python
import pandas as pd
import numpy as np
%load_ext rmagic
```

```python
counts = pd.read_csv("Parsing Friendly Tagger - Sheet1.tsv", sep="\t", header=None).rename(columns={0:"Id", 1:"POS", 2:"Counts"}).set_index("POS")["Counts"]
tags = counts[counts > 1000]
tags = set(tags.index)
```

```python
frame = pd.read_csv("Parsing Friendly Tagger - Large.tsv", sep="\t")
new_frame = frame.set_index("POS").stack().reset_index().rename(columns={"POS": "OTag", "level_1" : "RTag", 0: "Score"})
predicate = new_frame["OTag"].map(lambda x: x in tags) & new_frame["RTag"].map(lambda x: x in tags)
new_frame = new_frame[predicate]
None
```

```python
%%R -i new_frame
df <- as.data.frame(new_frame)
names(df) <- c("OTag","RTag","Score")
library(ggplot2)
p <- ggplot(new_frame, aes(x=OTag, y=RTag, fill=Score))
p <- p + 
    geom_tile() +
    ylab("Replaced Tag") +
    xlab("Original Tag") +
    ggtitle("Tag Replacements \n")
print(p)
```

```python
%%R -w 1500 -h 600 px
p <- ggplot(aes(x=RTag, y=Score), data=df)
p <- p + geom_bar(stat="identity") + ylab("Score") + xlab("Replaced")
p = p + facet_wrap(~OTag, scales="free")
print(p)
```

```python
%%R -w 1500 -h 600 px
p = ggplot(aes(x=factor(1), y=Score, fill=factor(RTag)), data=df)
p = p + geom_bar(stat="identity", width=10) + ylab("Score") + xlab("Replaced")
p = p + facet_wrap(~OTag, nrow=1, scales="free")
p = p + ggtitle("Faceted Replacements Dist\n")
print(p)
```
